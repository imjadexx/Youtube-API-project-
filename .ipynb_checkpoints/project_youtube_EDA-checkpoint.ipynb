```python
# Get channel stats
def get_channel_stats(youtube,channel_id):
    """
    Get from channel statistics: title, subscribe count,view count,videocount,uploads
    
    Parameters: the youtube object we create above , channel_id
    
    Return: Dataframe cotaining the channel statistics
    """
    
    all_data = []
    request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=channel_id
    )
    response_channel = request.execute()

    for item in response_channel['items']:
        data = {'channelName':item['snippet']['title'],
                'subscribers':item['statistics']['subscriberCount'],
                'views':item['statistics']['viewCount'],
                'totalVideos':item['statistics']['videoCount'],
                'playlistId':item['contentDetails']['relatedPlaylists']['uploads']
        }
        all_data.append(data)
        
    return pd.DataFrame(all_data)

get_channel_stats(youtube,channel_id)
```

```python
# Get Video ids in the list
playlist_id = 'UU2D2CMWXMOVWx7giW1n3LIg'

def get_video_ids(youtube, playlist_id):
    # Initialize variables
    video_ids = []
    next_page_token = None
    
    # Loop to fetch each page of data
    while True:
        request = youtube.playlistItems().list(
            part="contentDetails",
            playlistId=playlist_id,
            maxResults=50,
            pageToken=next_page_token
        )
        response_playlistitem = request.execute()

        for item in response_playlistitem['items']:
            video_ids.append(item['contentDetails']['videoId'])

        next_page_token = response_playlistitem.get('nextPageToken')
        
        # Break out of the loop if there are no more pages left
        if not next_page_token:
            break

    return video_ids
```

```python
# to check if it works
video_ids=get_video_ids(youtube, playlist_id)
len(video_ids)
```

```python
# Test
request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=video_ids[0:5])
response_video = request.execute()

JSON(response_video)    
```

```python
# Get video information for each video
def get_video_details(youtube, video_ids):
    
    all_video_info = [] # initialization: This list will hold dictionaries, with each dictionary containing details of a single video.
    
    for i in range(0,len(video_ids),50):
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_ids[i:i+50])) 
        response_video = request.execute()

        for video in response_video['items']:
            stats_to_keep = {'snippet':['channelTitle','title','description','tags','publishedAt'],
                             'statistics':['viewCount','likeCount','favoriteCount','commentCount'],
                             'contentDetails':['duration','definition','caption']
                            }
            video_info = {}
            
            video_info['video_id'] = video['id']# Ensures that the video ID is included in the details collected for each video.

            for k in stats_to_keep.keys(): # The nested loops iterate through the keys and values of stats_to_keep.
                                          
                for v in stats_to_keep[k]:
                    try:
                        video_info[v] = video[k][v]
                    except:
                        video_info[v] = None

            all_video_info.append(video_info) 
    
    return pd.DataFrame(all_video_info)
```

```python
video_df = get_video_details(youtube, video_ids)
```

### Title word cloud 

I'm also interested in what is the most frequently discussed topic of this podcast channel, we will plot word cloud to have a quick look of all titles. 
To do so we first need to remove all stop words('on', 'it', 'the' etc) which have no contribution to the theme. And because there is fixed words like 'Huberman', 'Lab', 'Podcast' in each title, we also need to include these words in our stop words.

```python
additional_stop_words = {'Huberman', 'Podcast','Lab','Dr'}
stop_words = set(stopwords.words('english')).union(additional_stop_words)
video_df['title_no_stopwords'] = video_df['title'].apply(lambda x: [item for item in str(x).split() if item not in stop_words])

all_words = list([a for b in video_df['title_no_stopwords'].tolist() for a in b])
text = ' '.join(all_words) 
```

```python
def plot_cloud(wordcloud):
    plt.figure(figsize=(30,20))
    plt.imshow(wordcloud)
    plt.axis("off");

wordcloud = WordCloud(width=2000,height=1000,random_state=1,colormap='Blues',collocations=False).generate(text)
    
plot_cloud(wordcloud)    
```

In the word cloud , we can find the most frequent words are Health, Science, Brain, Tool, which is very expected for a podcast about neuroscience.
